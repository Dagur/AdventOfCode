```python
# part 2
def check_n(x, y):
    i = x
    j = y - 1    
    while j >= 0:
        val = layout[x][j]
        if val == "#":
            return 1
        elif val == 'L':
            return 0
        j -= 1
    return 0

def check_ne(x, y):
    i = x + 1
    j = y - 1
    while j >= 0 and i < width:        
        val = layout[i][j]
        if val == "#":            
            return 1
        elif val == 'L':            
            return 0
        i += 1
        j -= 1    
    return 0

def check_e(x, y):
    i = x + 1    
    j = y
    while i < width:
        val = layout[i][y]
        if val == "#":
            return 1
        elif val == 'L':
            return 0
        i += 1
    return 0

def check_se(x, y):
    i = x + 1
    j = y + 1    
    while i < height and j < width:
        val = layout[i][j]
        if val == "#":
            return 1
        elif val == 'L':
            return 0
        i += 1
        j += 1

    return 0

def check_s(x, y):
    i = x
    j = y + 1  
    while j < width:
        val = layout[x][j]
        if val == "#":
            return 1
        elif val == 'L':
            return 0
        j += 1
    return 0

def check_sw(x, y):
    i = x - 1
    j = y + 1    
    while i >= 0 and j < width:        
        val = layout[i][j]
        if val == "#":            
            return 1
        elif val == 'L':            
            return 0
        i -= 1
        j += 
    return 0

    i = x - 1    
    j = y    
        val = layout[i][y]
        if val == "#":
            return 1
        elif val == 'L':
            return 0
        i -= 1
    return 0

def check_nw(x, y):
    i = x - 1
    j = y - 1    
    while i >= 0 and j >= 0:
        val = layout[i][j]
        if val == "#":
            return 1
        elif val == 'L':
            return 0
        i -= 1
        j -= 1

    return 0

changed = True
while changed:
    newlayout = [row.copy() for row in layout]
    changed = False
    for y in range(width):
        for x in range(height):
            seat = layout[x][y]        
            if seat == '.':
                continue
            occupied = \
                check_n(x,y) + \
                check_ne(x,y) + \
                check_e(x,y) + \
                check_se(x,y) + \
                check_s(x,y) + \
                check_sw(x,y) + \
                check_w(x,y) + \
                check_nw(x,y)
            if seat == 'L' and occupied == 0:
                changed = True
                newlayout[x][y] = "#"
            elif seat == '#' and occupied >= 5:
                changed = True                
                newlayout[x][y] = "L"
    layout = newlayout
    # for row in (''.join(fdaf) for fdaf in layout):
    #     print(row)
    # print()

print(sum(row.count("#") for row in layout))

```
